```python
df_country = pd.read_excel('/content/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code', how='left')
final_df.head()
```

```python
##To Check Data Types

final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts()
```

```python
country_values = final_df.Country.value_counts().values
```

```python
## Pie Chart

plt.pie(country_values, labels = country_names)
```

```python
## Pie Chart Top 3

plt.pie(country_values[:3], labels = country_names[:3],autopct='%1.2f%%')
```

Observations: Zomato maximum records or ransaction are from India Followed by USA and UK respectivelty

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

## Observations 

1. When Ratings are Between 4.5 to 4.9---> Excellent
2. When Ratings are Between 4.0 to 4.4--->Very Good
3. When Ratings are between 3.5 to 3.9---> Good
4. When Ratings are between 3.0 to 3.4---> Average
5. When Ratings are between 2.5 to 2.9---> Average
6. When Ratings are between 2.0 to 2.4---> Poor


```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data = ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue='Rating color', data = ratings, palette=['blue','red', 'orange','yellow','green', 'green' ])
```

Observations :
   1. Not Rated count is very high
   2. Maximum numbers of ratings are between 2.5 to 3.4
   

## Count Plot

```python


sns.countplot(x='Rating color',data = ratings, palette=['blue','red', 'orange','yellow','green', 'green' ])
```
